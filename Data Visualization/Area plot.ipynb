# Area plots

Es un tipo de gráfico que representa los totales acumulados utilizando números o
porcentajes en el tiempo. Se basa en el gráfico de líneas y se usa comúnmente cuando tratando de comparar dos o más cantidades. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'sales': [3, 2, 3, 9, 10, 6],
    'signups': [5, 5, 6, 12, 14, 13],
    'visits': [20, 42, 28, 62, 81, 50],
}, index=pd.date_range(start='2018/01/01', end='2018/07/01',
                       freq='M'))
ax = df.plot.area()
```
