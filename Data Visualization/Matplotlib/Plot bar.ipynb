# Plot bar

Es un diagrama que presenta datos categóricos con barras rectangulares con longitudes proporcionales a los valores que representan. Un gráfico de barras muestra comparaciones entre categorías discretas.

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'lab':['A', 'B', 'C'], 'val':[10, 30, 20]})
ax = df.plot.bar(x='lab', y='val', rot=0)
```

```python
speed = [0.1, 17.5, 40, 48, 52, 69, 88]
lifespan = [2, 8, 70, 1.5, 25, 12, 28]
index = ['snail', 'pig', 'elephant',
         'rabbit', 'giraffe', 'coyote', 'horse']
df = pd.DataFrame({'speed': speed,
                   'lifespan': lifespan}, index=index)
ax = df.plot.bar(rot=0)
```

```python
axes = df.plot.bar(
    rot=0, subplots=True, color={"speed": "red", "lifespan": "green"}
)
axes[1].legend(loc=2)  
```
